Seaborn visulization for the inbuilt data set - Tips

```python
import seaborn as sns
from IPython.display import display as dsp

dsp(sns.get_dataset_names())
```

```python
df = sns.load_dataset('tips')
dsp(df.head(10))
```

```python
df.info()
```

Distribution Plot

```python
sns.histplot(df['tip'], kde=True)
```

```python
sns.histplot(data=df, x='tip', element='poly', hue='smoker', kde=True,)
#hue is providinf data of various condition data famnes like sex, day, time etc
```

Joint Plot

```python
sns.jointplot(data=df, y='total_bill', x='smoker')
```

Pair Plot

```python
sns.pairplot(df, hue='smoker')
```

Strip Plot

```python
sns.stripplot(x='tip', y='time', data=df)
```

Box Plot

```python
#sns.boxenplot(data=df, x='tip', y='sex')
sns.boxplot(data=df, y='total_bill', x='day')
```
