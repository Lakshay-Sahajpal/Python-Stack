```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display as dsp

df = pd.read_csv(r'C:\Users\AKKODIS\Desktop\venv\code\cleaned_sales.csv')
df = df.set_index('OrderID')
dsp(df)
```

```python
df.plot()
```

```python
df.plot(subplots=True, title="Sales Data of Quantity, Price and Shipping Cost");
```

Setting individual y and x lebels

```python
axis = df.plot(subplots=True, layout=(3,1), figsize=(8,7), title="Sales Data of Quantity, Price and Shipping Cost")

#layout is no. of rows and columns in the plot)(3,1: 3 rows and 1 1 colum in each row)
#figsize is the size of indic=vidual subplots(width,hight)

axis[0][0].set_ylabel('Quantity')
axis[1][0].set_ylabel('Price(in $)')
axis[2][0].set_ylabel('Shippinp Cost(in $)')

plt.style.use('classic')
plt.tight_layout()
print(plt.style.available)
#or a just add y=['Quantity', 'Price($)', 'ShippingCost($)'] in the .plot()
```
